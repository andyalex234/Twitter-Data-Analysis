```python
#read csv 
clean_tweets_df = pd.read_csv('../clean_tweets.csv')
clean_tweets_df.head()
```

```python
# select only the columns we need
clean_tweets = clean_tweets_df[['created_at', 'original_text', 'polarity', 'subjectivity', 'retweet_count', 'favorite_count', 'lang']]
clean_tweets
```

```python
def text_category(p):
    """
    converts polarity into sentiment category
    """
    if p > 0:
        return "positive"
    elif p < 0:
        return "negative"
    else:
        return "neutral"
```

```python
clean_tweets["score"] = clean_tweets["polarity"].apply(text_category)
clean_tweets
```

#### Sentiment Analysis

```python
sentiment = clean_tweets['score'].value_counts()
sentiment
```

#### After Removing neutral sentiments

```python
sentiment = clean_tweets[clean_tweets['score'] != 'neutral'].groupby('score').count()
sentiment
```
